Loading gapminder data for year 1985 (Live Aid!) and setting up plot as in chapter 3

```python
gapminder = pd.read_csv('gapminder.csv')
```

```python
gdata = gapminder.query('year == 1985')
```

```python
size = 1e-6 * gdata.population

colors = gdata.region.map({'Africa': 'skyblue', 'Europe': 'gold', 'America': 'palegreen', 'Asia': 'coral'})

def plotdata():
    gdata.plot.scatter('age5_surviving','babies_per_woman',
                      c=colors,s=size,linewidths=0.5,edgecolor='k',alpha=0.5)
```

Setting up model plot

```python
def plotfit(fit):
    plotdata()
    pp.scatter(gdata.age5_surviving,fit.predict(gdata),
               c=colors,s=30,linewidths=0.5,edgecolor='k',marker='D')
```

Three models from last video

```python
groupmeans = smf.ols(formula='babies_per_woman ~ -1 + region', data=gdata).fit()
```

```python
surviving = smf.ols(formula='babies_per_woman ~ -1 + region + age5_surviving', data=gdata).fit()
```

```python
surviving_byregion_population = smf.ols(
    formula='babies_per_woman ~ -1 + region + age5_surviving'
            '+ age5_surviving:region - age5_surviving + population',
    data=gdata).fit()
```

```python
plotfit(surviving)
```

```python
plotfit(surviving_byregion_population)
```

```python
# Scatterplot with residuals
# simple linear regression assumes that a linear relationship exists between independent and dependent variables
# residual analysis tests that assumption
# residual is observed - predicted, residuals are plotted on the y-axis and the independent variable on the x-axis
# if pattern is random, linear regression is appropriate, otherwise not appropriate.
pp.scatter(gdata.age5_surviving,groupmeans.resid)
pp.scatter(gdata.age5_surviving,surviving.resid)
pp.scatter(gdata.age5_surviving,surviving_byregion_population.resid)
```

```python
# fit results object contains the mean square residual, which gives goodness of fit in single number,
# the mean squared error for the three models
for model in [groupmeans, surviving, surviving_byregion_population]:
    print (model.mse_resid)
```

```python
# from these results we can see that one model is slightly better fit
# rsquared also good for establishing goodness of fit, R^2 = (explainedvariance)/(totalvariance)
# rquared values vary from 0(min) to 1(max)
for model in [groupmeans, surviving, surviving_byregion_population]:
    print (model.rsquared)
```

```python
#Statsmodels also includes the F statistic, which takes into account variance and number of parameters.
# Too many variables may overfit, losing predictive powers. F value should show how much each variable contributes
# to the rsquared error
```

```python
for model in [groupmeans, surviving, surviving_byregion_population]:
    print (model.fvalue)
```
